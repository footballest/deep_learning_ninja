
 # Reshaping, viewing, stackign, squeeze and unqueeze, permute



```python
x = torch.arange(1., 10.)
x, x.shape
```

```python
x_reshaped = x.reshape(len(x), -1)
x_reshaped, x_reshaped.shape
```

```python
z = x.view(1, 9)
z, z.shape
```

```python
x.view(1, 9)
```

```python
# z shares the same memory as x: i.e the view of a tensor shares the same memory
z[:, 0] = 5
z, x
```

```python
x_stacked = torch.stack([x, x, x, x], dim=1)
x_stacked
```

```python
z.shape
```

```python
y = torch.rand(3,4)
torch.squeeze(y)
```

```python
x_reshaped
```

```python
x_squeezed = x_reshaped.squeeze() ## squeeze removes all single dimensions in a tensor
x_squeezed, x_squeezed.shape
```

```python
## torch.unsqueeze() adds a single dimension to a target tensor at a specific dim
x_unsqueezed = x_squeezed.unsqueeze(dim=-1)
x_unsqueezed, x_unsqueezed.shape
```

```python
## torch.permute() rearranges the dimensions of a tensor to a specific shape - returns a view
x_original = torch.rand(size=(224, 224, 3))
x_permuted = x_original.permute(2, 0, 1)
x_original.shape, x_permuted.shape
```
