```python
A = torch.randn(100, 200)
B = torch.randn(100, 200)
C = torch.randn(100, 200)

X = torch.stack([A, B, C], dim=2)   # shape (100, 200, 3)
B_recovered = X[:, :, 1]            # shape (100, 200)

X.shape, torch.allclose(B, B_recovered)      # True
```

```python
A = torch.randn(100, 200)
B = torch.randn(100, 200)
C = torch.randn(100, 200)

X = torch.stack([A, B, C], dim=0)   # shape (100, 200, 3)
B_recovered = X[1, :, :]            # shape (100, 200)

X.shape, torch.allclose(B, B_recovered)     # True
```
