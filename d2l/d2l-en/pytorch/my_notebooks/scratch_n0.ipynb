## broadcasting:  principles
1. align the two tensors, copy either of the tensors, if possible, along axis with 1 to match the other tensor, if axis misssing - insert new axis and expand
2. perform an element wise operation after matching the dimensions.
3. Along each trailing dimension, the dimension sizes must either be: (1) equal, (2) one of them is 1, or (3) one of them does not exist.
